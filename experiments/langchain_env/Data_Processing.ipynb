### 3-1 RecursiveCharacterTextSplitter

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,        # ì²­í¬ í¬ê¸°  
    # ë¬¸ì¥ì´ ì¤‘ê°„ì— ë‚˜ëˆ ì§€ë”ë¼ë„, ì¼ì •ë¶€ë¶„ ê³‚ì¹˜ë„ë¡ í•´ì„œ
    # ì •ë³´ì˜ ì†Œì‹¤ì—†ì´ ë¬¸ë§¥ì´ ì˜ ë³´ì¡´ë˜ë„ë¡ ì–´ëŠì •ë„ì˜ chunk overlapì„ ì§€ì •
    chunk_overlap=200,      # ì²­í¬ ì¤‘ ì¤‘ë³µë˜ëŠ” ë¶€ë¶„ í¬ê¸°
    length_function=len,    # ê¸€ì ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„í• 
    # separators=["\n\n"],  # êµ¬ë¶„ìë¥¼ ëœ ì—„ê²©í•˜ê²Œ ì§€ì •í•  ìˆ˜ë¡ chunk ì‚¬ì´ì¦ˆê°€ ì˜ ì§€ì¼œì§€ì§€ ì•Šì„ ìˆ˜ ìˆìŒ
    separators=["\n\n", "\n", ".", " ", ""],  # ì—„ê²©í•˜ê²Œ ì²­í¬ì‚¬ì´ì¦ˆë¥¼ ì§€ì¼œì•¼ í•œë‹¤ë©´ êµ¬ë¶„ìë¥¼ ì—„ê²©í•˜ê²Œ ì„¤ì •
)

texts = text_splitter.split_documents(pdf_docs)
print(f"ìƒì„±ëœ í…ìŠ¤íŠ¸ ì²­í¬ ìˆ˜: {len(texts)}")
print(f"ê° ì²­í¬ì˜ ê¸¸ì´: {list(len(text.page_content) for text in texts)}")
````

- RecursiveCharacterTextSplitterëŠ” ì´ë¦„ì—ì„œ ì•Œ ìˆ˜ ìˆë“¯ì´ ì¬ê·€ì ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•©ë‹ˆë‹¤. 
- êµ¬ë¶„ìë¥¼ ìˆœì°¨ì ìœ¼ë¡œ ì ìš©í•˜ì—¬ í° ì²­í¬ì—ì„œ ì‹œì‘í•˜ì—¬ ì ì§„ì ìœ¼ë¡œ ë” ì‘ì€ ë‹¨ìœ„ë¡œ ë‚˜ëˆ•ë‹ˆë‹¤.

### 3-2 ì •ê·œí‘œí˜„ì‹ ì‚¬ìš©

````python
# ë¬¸ì¥ì„ êµ¬ë¶„í•˜ì—¬ ë¶„í•  (ë§ˆì¹¨í‘œ, ëŠë‚Œí‘œ, ë¬¼ìŒí‘œ ë‹¤ìŒì— ê³µë°±ì´ ì˜¤ëŠ” ê²½ìš° ë¬¸ì¥ì˜ ëìœ¼ë¡œ íŒë‹¨)
from langchain_text_splitters import CharacterTextSplitter

# ì •ê·œì‹ Seperatorë¥¼ ì¬ê·€ì ìœ¼ë¡œ ì ìš©í•  ê²ƒì´ë¼ë©´ 
# RecursiveCharacterTextSplitter ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°°ì—´ë¡œ separatorì— ì •ê·œì‹ì„ ì‚¬ìš©í•  ìˆ˜ë„ ìˆìŒ.
text_splitter = CharacterTextSplitter(
    chunk_size=10,
    chunk_overlap=0,
    separator=r'(?<=[.!?])\s+',     # (ë§ˆì¹¨í‘œ or ëŠë‚Œí‘œ or ë¬¼ìŒí‘œ) ë‹¤ìŒì— ê³µë°±ì´ ì˜¤ëŠ” ê²½ìš° ë¬¸ì¥ì˜ ëìœ¼ë¡œ íŒë‹¨
    is_separator_regex=True,        # ì •ê·œì‹ ì—¬ë¶€
    keep_separator=True,
)

texts = text_splitter.split_documents(json_docs)
print(f"ìƒì„±ëœ í…ìŠ¤íŠ¸ ì²­í¬ ìˆ˜: {len(texts)}")
print(f"ê° ì²­í¬ì˜ ê¸¸ì´: {list(len(text.page_content) for text in texts)}")
print()

# ê° ì²­í¬ì˜ ì‹œì‘ ë¶€ë¶„ê³¼ ë ë¶€ë¶„ í™•ì¸
for text in texts[:5]:
    print(text.page_content[:50])
    print("-" * 50)
    print(text.page_content[-50:])
    print("=" * 50)
    print()
````

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",        # í•´ë‹¹ ì¸ì½”ë”© ì´ë¦„ì„ ì‚¬ìš©
    # model_name="gpt-4o-mini",         # í•´ë‹¹ ëª¨ë¸ì´ ì‚¬ìš©í•˜ëŠ” tokenizerë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆìŒ
    chunk_size=300,                     # tiktoken - chunkì˜ í¬ê¸°ë¼ê¸° ë³´ë‹¤ëŠ” í† í°ì˜ ìˆ˜
    chunk_overlap=0,
)

chunks = text_splitter.split_documents(pdf_docs[:1]) #ìŠ¬ë¼ì´ì‹±ì„ ì‚¬ìš©í•´ì„œ ì²«ë²ˆì§¸ ìš”ì†Œ ê¹Œì§€ë§Œ ê°€ì§„ ë°°ì—´ì„ ë¦¬í„´

print(f"ìƒì„±ëœ ì²­í¬ ìˆ˜: {len(chunks)}")
print(f"ê° ì²­í¬ì˜ ê¸¸ì´: {list(len(chunk.page_content) for chunk in chunks)}")

# ê° ì²­í¬ì˜ ì‹œì‘ ë¶€ë¶„ê³¼ ë ë¶€ë¶„ í™•ì¸
for chunk in chunks[:5]:
    print(chunk.page_content[:50])
    print("-" * 50)
    print(chunk.page_content[-50:])
    print("=" * 50)
    print()
````

````python
import tiktoken

tokenizer = tiktoken.get_encoding("cl100k_base")
# tokenizer = tiktoken.encoding_for_model("gpt-4o-mini")

for chunk in chunks[:5]:

    # ê° ì²­í¬ë¥¼ í† í°í™”
    tokens = tokenizer.encode(chunk.page_content)
    # ê° ì²­í¬ì˜ ë‹¨ì–´ ìˆ˜ í™•ì¸
    print(len(tokens))
    # ê° ì²­í¬ì˜ í† í°í™” ê²°ê³¼ í™•ì¸ (ì²« 10ê°œ í† í°ë§Œ ì¶œë ¥)
    print(tokens[:10])
    # í† í° IDë¥¼ ì‹¤ì œ í† í°(ë¬¸ìì—´)ë¡œ ë³€í™˜í•´ì„œ ì¶œë ¥
    token_strings = [tokenizer.decode([token]) for token in tokens[:10]]
    print(token_strings)

    print("=" * 50)
    print()
````

````python
from langchain_community.document_loaders import TextLoader

def load_text_files(txt_files):
    data = []

    for text_file in txt_files:
        loader = TextLoader(text_file, encoding='utf-8')
        data += loader.load()

    return data


korean_data = load_text_files(korean_txt_files)
english_data = load_text_files(english_txt_files)

print("í•œêµ­ì–´ ë°ì´í„° ìˆ˜:", len(korean_data))
print("ì˜ì–´ ë°ì´í„° ìˆ˜:", len(english_data))
````

`(3) RAG ì„±ëŠ¥ ë¹„êµ `  

````python
# RAG ì²´ì¸ ìƒì„±
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnablePassthrough

template = """Answer the question based only on the following context.
Do not use any external information or knowledge. 
If the answer is not in the context, answer "ì˜ ëª¨ë¥´ê² ìŠµë‹ˆë‹¤.".

[Context]
{context}

[Question] 
{question}

[Answer]
"""

prompt = ChatPromptTemplate.from_template(template)


# ë¬¸ì„œ í¬ë§·í„° í•¨ìˆ˜
def format_docs(docs):
    return "\n\n".join([d.page_content for d in docs])


# LLM ëª¨ë¸ ìƒì„±
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# ì²´ì¸ ìƒì„±
def create_rag_chain(vectorstore):

    retriever = vectorstore.as_retriever(search_kwargs={'k': 2})

    return (
        {"context": retriever | format_docs , "question": RunnablePassthrough()}
        | prompt
        | llm
        | StrOutputParser()
    )
````

````python
# ì²´ì¸ ìƒì„±
rag_chain_openai = create_rag_chain(db_openai)
rag_chain_huggingface = create_rag_chain(db_huggingface)
rag_chain_ollama = create_rag_chain(db_ollama)
````

### 5-2 ì–¸ì–´ ê°ì§€ ë° ìë™ë²ˆì—­ í†µí•©

````python
# í•œêµ­ì–´ ë¬¸ì„œë¡œ ì €ì¥ë˜ì–´ ìˆëŠ” ë²¡í„° ì €ì¥ì†Œ ë¡œë“œ
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small", 
)

vectorstore = Chroma(
    embedding_function=embeddings,
    collection_name="chroma_test",
    persist_directory="./chroma_db",
    )

print(f"ë²¡í„° ì €ì¥ì†Œì— ì €ì¥ëœ ë¬¸ì„œ ìˆ˜: {vectorstore._collection.count()}")
````

````python
import deepl
from langdetect import detect

translator = deepl.Translator(os.getenv('DEEPL_API_KEY'))

def detect_and_translate(text, target_lang='KO'):
    detected_lang = detect(text)
    if detected_lang.upper() != target_lang:
        result = translator.translate_text(text, target_lang=target_lang)
        return str(result), detected_lang
    return text, detected_lang

# ë¬¸ì„œ ë²ˆì—­ í…ŒìŠ¤íŠ¸
text = "Hello. How are you?"
translated_text, detected_lang = detect_and_translate(text, target_lang='KO')
print(f"Detected language: {detected_lang}")
print(f"Translated text: {translated_text}")
````

`(2) RAG ì²´ì¸ ì„±ëŠ¥ í‰ê°€ `  

````python
retriever = vectorstore.as_retriever(search_kwargs={'k': 2})

lang_rag_chain = (
        {"context": retriever | format_docs , "question": RunnablePassthrough()}
        | prompt
        | llm
        | StrOutputParser()
    )

# ì²´ì¸ ì‹¤í–‰ í•¨ìˆ˜
def run_lang_rag_chain(query):
    original_lang = detect(query)
    
    if original_lang.upper() != 'KO':
        translated_query, _ = detect_and_translate(query, target_lang='KO')
    else:
        translated_query = query
    
    output = lang_rag_chain.invoke(translated_query)
    
    if original_lang.upper() != 'KO':
        target_lang = 'EN-US' if original_lang.upper() == 'EN' else original_lang
        output, _ = detect_and_translate(output, target_lang=target_lang)
    
    return output
    
````

````python
# í•œêµ­ì–´ ì¿¼ë¦¬ì— ëŒ€í•œ ì„±ëŠ¥ í‰ê°€
query_ko = "í…ŒìŠ¬ë¼ ì°½ì—…ìëŠ” ëˆ„êµ¬ì¸ê°€ìš”?"

output = run_lang_rag_chain(query_ko)
print(output)

# ì˜ì–´ ì¿¼ë¦¬ì— ëŒ€í•œ ì„±ëŠ¥ í‰ê°€
query_en = "Who is the founder of Tesla?"

output = run_lang_rag_chain(query_en)
print(output)
````

````python
import gradio as gr
from langchain_core.messages import HumanMessage, AIMessage

# ë‹µë³€ ìƒì„± ëª¨ë¸ì„ ë³„ë„ë¡œ ì‚¬ìš©
answer_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)


def answer_invoke(message, history):
    history_langchain_format = []
    for human, ai in history:
        history_langchain_format.append(HumanMessage(content=human))
        history_langchain_format.append(AIMessage(content=ai))
    history_langchain_format.append(HumanMessage(content=message))

    # í˜„ì¬ ë©”ì‹œì§€ì— ëŒ€í•´ RAG ì²´ì¸ ì‹¤í–‰
    rag_response = run_route_rag_chain(message)

    # ë‹µë³€ ìƒì„± ëª¨ë¸ì—ê²Œ í˜„ì¬ ë©”ì‹œì§€ì— ëŒ€í•œ ë‹µë³€ ìš”ì²­
    final_answer = answer_llm.invoke(
        history_langchain_format[:-1] + [AIMessage(content=rag_response)] + [HumanMessage(content=message)]
    )
    
    return final_answer.content


# Graiio ì¸í„°í˜ì´ìŠ¤ ìƒì„± 
demo = gr.ChatInterface(fn=answer_invoke, title="QA Bot")

# Graiio ì‹¤í–‰  
demo.launch()
````
